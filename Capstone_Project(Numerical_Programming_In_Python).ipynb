#### Chart - 14 - Correlation Heatmap

Now I will convert our main column that is Churn column into binary that is 0 and 1 form for our heatmap to check the correlation between the columns and false will give 0 and true will give 1..

````python
df2=pd.get_dummies(df["Churn"])
df2.head()
````

````python
df3=pd.concat((df,df2),axis=1)
df3.head()
````

````python
df3=df3.drop(["Churn",False],axis=1)
````

````python
df3=df3.rename(columns={True:"Churn"})
````

````python
# Correlation Heatmap visualization code
df_heatmap=df3[["Account length",	"Area code",	"Number vmail messages",	"Total day minutes",	"Total day calls",	"Total day charge",	"Total eve minutes",	"Total eve calls",	"Total eve charge",	"Total night minutes",	"Total night calls",	"Total night charge",	"Total intl minutes",	"Total intl calls",	"Total intl charge",	"Customer service calls","Churn"]]
plt.figure(figsize=(18,8))
sns.heatmap(df_heatmap.corr(), cmap="YlGnBu", annot=True)
plt.show()
````

##### 1. Why did you pick the specific chart?

It has been taken to check correlation between the columns that how they are related to each other that is either one is increasing and other also increasing or one increasing and other decreasing..

##### 2. What is/are the insight(s) found from the chart?

There are columns which have correlation of 1 also with another column like we can see above that there is correlation 1 of total day charge with total day minutes by which we can know that if total day minutes are increasing than total day charge is also increasing which tells that our data is also clean.
